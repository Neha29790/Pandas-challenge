```python
# Dependencies and Setup
import pandas as pd
from babel.numbers import format_currency

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete_df = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## Local Government Area Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average maths score 

* Calculate the average reading score

* Calculate the percentage of students with a passing maths score (50 or greater)

* Calculate the percentage of students with a passing reading score (50 or greater)

* Calculate the percentage of students who passed maths **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
TotalSchools= len(school_data_complete_df.school_name.unique())

TotalStudents= len(school_data_complete_df.student_name)

TotalBudget= school_data.budget.sum()

TotalAvgMathScore= school_data_complete_df.maths_score.mean()

TotalAvgReadingScore= school_data_complete_df.reading_score.mean()

TotalStudentSWith50MathScore = school_data_complete_df.apply(lambda x : True
            if x['maths_score'] >= 50 else False, axis = 1)

num_rows = len(TotalStudentSWith50MathScore[TotalStudentSWith50MathScore == True].index)
TotalStudentPercentageWith50MathScore = (num_rows/TotalStudents)*100

TotalStudentSWith50ReadScore = school_data_complete_df.apply(lambda x : True
            if x['reading_score'] >= 50 else False, axis = 1)

num_rows = len(TotalStudentSWith50ReadScore[TotalStudentSWith50ReadScore == True].index)
TotalStudentPercentageWith50ReadScore= (num_rows/TotalStudents)*100

TotalStudentSWithPassedInBoth = school_data_complete_df.apply(lambda x : True
            if x['reading_score'] >= 50 and x['maths_score'] >= 50 else False, axis = 1)

num_rows = len(TotalStudentSWithPassedInBoth[TotalStudentSWithPassedInBoth == True].index)
TotalStudentPercentageOverallPassing =(num_rows/TotalStudents)*100

d = {'Total Schools': [TotalSchools], 'Total Students': [TotalStudents], 'Total Budget': [TotalBudget], 'Average maths score': [TotalAvgMathScore], 'Average reading score': [TotalAvgReadingScore], '% passing maths': [TotalStudentPercentageWith50MathScore], '% passing reading': [TotalStudentPercentageWith50ReadScore], '% overall passing': [TotalStudentPercentageOverallPassing]}
df = pd.DataFrame(data=d)
df['Total Schools'] = df['Total Schools'].map('{:,d}'.format)
df['Total Students'] = df['Total Students'].map('{:,d}'.format)
df['Total Budget'] = df['Total Budget'].apply(lambda x: format_currency(x, currency="AUD", locale="en_AU"))
df
```

## School Summary

* Create an overview table that summarises key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed maths **and** reading.)
  
* Create a dataframe to hold the above results

```python
#school_data_complete_df.reset_index(inplace=True, drop=True)

school_type = school_data.set_index(["school_name"])["type"]

size = school_data.set_index(["school_name"])["size"]

budget = school_data.set_index(["school_name"])["budget"]

number_of_students = school_data_complete_df["school_name"].value_counts()

each_student_budget =budget / number_of_students

maths_score = school_data_complete_df.groupby(["school_name"])["maths_score"].mean()

reading_score = school_data_complete_df.groupby(["school_name"])["reading_score"].mean()

passing_math = school_data_complete_df[(school_data_complete_df["maths_score"] >= 50)]

students_passing_math = passing_math.groupby(["school_name"]).count()["student_name"]

passing_read = school_data_complete_df[(school_data_complete_df["reading_score"] >= 50)]

students_passing_read = passing_read.groupby(["school_name"]).count()["student_name"]

passing_math_percentage = students_passing_math /number_of_students * 100

passing_read_percentage = students_passing_read /number_of_students * 100

total_passing = school_data_complete_df[(school_data_complete_df["maths_score"] >= 50) & (school_data_complete_df["reading_score"] >= 50)]

total_passing_percentage = total_passing.groupby(["school_name"]).count()["student_name"]

per_overall_passing_percentage = total_passing_percentage / number_of_students *100

school_summary_df = pd.DataFrame({
                
                "School Type": school_type,
                "Total Students": size,
                "Total School Budget": budget,
                "Per Student Budget": each_student_budget,
                "Average Math Score": maths_score,
                "Average Reading Score": reading_score,
                "% Passing Math": passing_math_percentage,
                "% Passing Reading": passing_read_percentage,
                "% Overall Passing": per_overall_passing_percentage})
school_summary_df
```

```python
top_schools = school_summary_df.sort_values(["% Overall Passing"], ascending=False)
top_schools.head()
```

## Maths Scores by Year

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
spending_bins = [0, 585, 630, 645, 680]
group_names = ["<$584", "$585-630", "$630-645", "$645-680"]

school_summary_df["Spending Ranges (Per Student)"] = pd.cut(each_student_budget, spending_bins, labels=group_names)
school_summary_df.head()

spending_math_scores = school_summary_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average Math Score"]

spending_reading_scores = school_summary_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average Reading Score"]

spending_passing_math = school_summary_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Math"]

spending_passing_reading = school_summary_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Reading"]

overall_passing_spending = school_summary_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Overall Passing"]

spending_summary_df = pd.DataFrame({
                "Average Math Score": spending_math_scores,
                "Average Reading Score": spending_reading_scores,
                "% Passing Math": spending_passing_math,
                "% Passing Reading": spending_passing_reading,
                "% Overall Passing": overall_passing_spending })
spending_summary_df
```
